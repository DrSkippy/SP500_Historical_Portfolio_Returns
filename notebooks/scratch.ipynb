```python
def plot_return_comps(frac_list = [], trend_list = [], year=10, index=3):
    a = [x[year-1] for x in trend_list]
    b = np.array(a).T[:index+1]
    b = b.astype(float)
    plt.plot(frac_list, b[index])
    plt.legend()
    plt.show()
```
